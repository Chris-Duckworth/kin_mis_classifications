# create_dataset  

Creating velocity field datasets for NN purposes.

--- 

- X data is found from raw MaNGA data (DAP output) and pixel values are scaled to [-1, 1] range and the image is resized to [32, 32]. (Currently the images are not recentred (i.e. the velocity field centre doesn't necessarily match the centre of the image), however, this may actually be an indication of whether you can trust the vel field anyway so leaving for now)

- y data is found from the expert (human) classifications of velocity fields so that: 
    - 0 - Do not use
    - 1 - Messy (potentially usable but maps are typically messy and global PAs will have higher errors)
    - 2 - Clean (global PAs are well behaved and well defined)

- X and y are saved as np.arrays for stars and gas individually and also combined (to increase sample size)

```python
import pandas as pd
```

```python
class_dir = '/Users/chrisduckworth/projects/ML_portfolio/kin_mis_classification/expert_classifications/current_release/'
tab = pd.read_csv(class_dir + 'mpl8-pa-classifications-v0_1.csv')
tab
```

```python
# write separate functions to build stars and gas classifications...?

# defining empty X and y.

X_stel = []
y_stel = []

X_gas = []
y_gas = []

for ind, pifu in enumerate(progressbar.progressbar(tab.plateifu.values)):
    # grabbing processed velocity fields, if file doesn't exist continuing
    try:
        stel, gas = data_prep.processed_vfields(pifu, data_dir, shape=(32, 32))
    except:
        continue
    
    X_stel.append(stel) ; X_gas.append(gas)
    
    # adding flags
    y_stel.append(tab.stel_qual.values[ind])
    y_gas.append(tab.halpha_qual.values[ind])
```

```python
# converting to numpy arrays and saving for future use.
outpath = '/Users/chrisduckworth/projects/ML_portfolio/kin_mis_classification/CNN/pre-processing/raw_dataset_files/'

np.save(outpath + 'X_stel', X_stel)
np.save(outpath + 'X_gas', X_gas)
np.save(outpath + 'y_stel', y_stel)
np.save(outpath + 'y_gas', y_gas)

np.save(outpath + 'X_total', np.concatenate((X_stel, X_gas)))
np.save(outpath + 'y_total', np.concatenate((y_stel, y_gas)))
```
