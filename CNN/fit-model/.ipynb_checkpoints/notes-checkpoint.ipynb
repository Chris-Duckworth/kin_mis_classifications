```python
import keras 
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from keras import Sequential
from keras.layers import Conv2D, Flatten, BatchNormalization, Dropout, MaxPool2D, Dense
from sklearn.utils import class_weight

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Loading in pre-processed data

```python
# see ../pre-processing/clean_dataset/ for more info
filepath = '/Users/chrisduckworth/projects/ML_portfolio/kin_mis_classification/CNN/pre-processing/clean_dataset_files/'

X_train = np.load(filepath + 'X_train.npy')
X_test = np.load(filepath + 'X_test.npy')
X_val = np.load(filepath + 'X_val.npy')

y_train = np.load(filepath + 'y_train.npy')
y_test = np.load(filepath + 'y_test.npy')
y_val = np.load(filepath + 'y_val.npy')
```

```python
class_weights = class_weight.compute_class_weight('balanced', 
                                                  np.unique(y_train), 
                                                  y_train)
weight = {i : class_weights[i] for i in range(3)}
```

### Reshaping X data (i.e. adding channel) for CNN input.

```python
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
X_val = X_val[..., np.newaxis]
```

### Our data is mildly imbalanced, so creating class weights using sklearn

```python
### Updating labels to categorical

y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
y_val = keras.utils.to_categorical(y_val)
```

### Using data generator objects to automatically augment training images to increase sample size.

```python
train_gen = ImageDataGenerator(rotation_range = 45,  # randomly rotate images in the range (degrees, 0 to 180)
                               horizontal_flip = True, 
                               vertical_flip = True)  # randomly flip images

train_gen.fit(X_train)

valid_gen = ImageDataGenerator()
valid_gen.fit(X_val)
```

```python
def build_seq_CNN(input_shape):
    '''
    Building CNN model using keras sequential API.
    '''
    model = Sequential()
    
    # First convolutional block
    model.add(Conv2D(32 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = input_shape))
    # Batch normalisation normalises output between layers 
    # (in a similar way to that data is normalised between 0 and 1 in the input layer).
    # This enables higher learning rates / quicker convergence.
    model.add(BatchNormalization())
    # Pooling to avert small positional changes in features.
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))

    # Second convolutional block
    model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
    model.add(BatchNormalization())
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))

    # Third convolutional block
    model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
    model.add(BatchNormalization())
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))
    
    # Fully connect layer and output.
    model.add(Flatten())
    model.add(Dense(units = 128 , activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units = 3 , activation = 'softmax')) # 3 classes for output

    # Compiling model with binary crossentropy. 
    model.compile(optimizer = "adam" , loss = 'categorical_crossentropy' , metrics = ['accuracy'])

    return model


def build_seq_FCN():
    '''
    Building simple sequential model of fully connected neural network.    
    '''
    
    # initiating sequential model object.
    model = keras.models.Sequential()

    # flattening 28 by 28 input image to single vector.
    # note that if the flattening is not used, then an input layer (shape needs to be defined) must be specified.
    model.add(Flatten(input_shape=[32, 32, 1]))
    
    # adding layer of fully connected neurons with a relu activation function.
    # the input layer is connected to 256 output neurons
    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.2))
    
    # adding further layer of FCN with relu activation function.
    model.add(Dense(128, activation="relu"))
    
    # adding further layer of FCN with relu activation function.
    model.add(Dense(256, activation="relu"))
    
        # adding further layer of FCN with relu activation function.
    model.add(Dense(128, activation="relu"))

    # adding final layer of FCN with softmax activation function.
    # softmax is often used as the last activation function of a neural network to normalize the output 
    # of a network to a probability distribution over predicted output classes.
    model.add(Dense(3, activation="softmax"))
    
    model.compile(optimizer = "sgd" , loss = 'categorical_crossentropy' , metrics = ['accuracy'])

    return model
```

```python
model = build_seq_FCN()
model.summary()
```

```python
model.fit(train_gen.flow(X_train, y_train, batch_size=32),
          validation_data=valid_gen.flow(X_val, y_val), 
          class_weight=weight, epochs=50)
```

```python
plot_loss_acc(model.history)
```

```python
y_predict_classes = model.predict_classes(X_test)
y_predict = keras.utils.to_categorical(y_predict_classes)
y_test_classes = np.argmax(y_test, axis=1)

plt.hist(y_test_classes, label='Actual')
plt.hist(y_predict_classes, alpha=0.6, label='Predictions')
plt.legend()
```

```python
plot_confusion_matrix(y_test_classes, y_predict_classes, labels=['dont', 'maybe', 'clean'])
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import harry_plotter as hp
from sklearn.metrics import classification_report, confusion_matrix

# updating default plotting style.
hp.default(plt.rcParams)


def plot_loss_acc(history):
	'''
	Given model training history (output from keras.model.fit), this returns a quick plot
	containing training and validation loss and accuracy for comparison.
	'''
	# init figure
	fig, ax = plt.subplots(1,2, figsize=(10, 4), sharex='row')

	# loss
	ax[0].set_title('Loss', fontsize=14)
	ax[0].plot(history.history['loss'], 'salmon', label='training')
	ax[0].plot(history.history['val_loss'], 'navy', label='validation')
	ax[0].legend(frameon=False, fontsize=12)
	ax[0].set_xlabel('Epoch', fontsize=12)
	hp.xtick_format(5, 1, ax=ax[0], format='%1.0f')

	# accuracy
	ax[1].set_title('Accuracy', fontsize=14)
	ax[1].plot(history.history['accuracy'], 'salmon', label='training')
	ax[1].plot(history.history['val_accuracy'], 'navy', label='validation')
	ax[1].legend(frameon=False, fontsize=12)
	ax[1].set_xlabel('Epoch', fontsize=12)
	
	return ax


def plot_confusion_matrix(y_test, y_predict_classes, labels):
	'''
	Quick (binary) confusion matrix.
	'''
	
	# sklearn confusion matrix
	con_mat = confusion_matrix(y_test, y_predict_classes)
	
	# plotting using seaborn heatmap, also can use plt.matshow
	fig = plt.figure(figsize=(8,8))
	ax = fig.add_subplot()
	sns.set(font_scale=1) # for label size
	sns.heatmap(con_mat, annot=True, fmt='d', cmap='Blues', 
				xticklabels=labels, yticklabels=labels) 

	ax.set_xlabel('Actual Value', fontsize=15)
	ax.set_ylabel('Model Prediction', fontsize=15)

	ax.set_xlim([0, 3])
	ax.set_ylim([0, 3])
	return ax
```
