```python
# loading in new dataframe with comments
tab = pd.read_csv('/Users/cd201/mpl8-kin-mis/classification/release-1/mpl8-pa-classifications.csv', comment='#')
```

```python
# Adding pa offset column.
def pa_offset(spa, gpa):
    off_cw = np.abs(spa - gpa)
    off_ccw = 360-np.abs(spa - gpa)
    pa_offset = np.minimum(off_cw, off_ccw)
    return pa_offset
```

```python
tab['pa_offset'] = pa_offset(tab.stellar_pa.values, tab.halpha_pa.values)
```

```python
tab = tab.rename(index=str, columns={"STEL_QUAL": "stel_qual", "GAS_QUAL": "halpha_qual", "STEL_FEATURE":'stel_feature', 'GAS_FEATURE':'halpha_feature', "GAS_NOTES":'halpha_notes'})
tab
```

```python
# Reordering columns
tab = tab[['plateifu','stellar_pa','halpha_pa','pa_offset','stel_qual','stel_feature','halpha_qual','halpha_feature','halpha_notes']]
```

```python
# This apparently wipes out the comment lines. re-adding.
# Adding additional comment lines.
f = open('mpl8-pa-classifications-v0_1.csv', 'a')
f.write('# mpl8-pa-classifications: 17-4-19. v0.1 - still in vetting.\n')
f.write('# Chris Duckworth. cd201@st-andrews.ac.uk\n')
tab.to_csv(f, index=None)
f.close()
```

```python
# Checking classifications given.
clean = tab[(tab.stel_qual == 2) & (tab.halpha_qual == 2) & (tab.stel_feature == 0) & (tab.halpha_feature == 0)]
messy = tab[(tab.stel_qual == 1) & (tab.halpha_qual == 1) & (tab.stel_feature == 0) & (tab.halpha_feature == 0)]
```

```python
# concatenating arrays.
clean_messy = clean.append(messy)
```
